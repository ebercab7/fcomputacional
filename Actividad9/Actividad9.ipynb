# **Ejercicio 1:**


\begin{eqnarray*}
\frac{d x}{d t} & = & y  \\
\frac{d y}{dt } & = & -x   \\
\end{eqnarray*}

```python
# Se definen a, b, c, d:
a = 0
b = 1
c = -1
d = 0

# Se define la matriz A
A = np.array([[a, b],[c, d]])

# Se obtienen los eigenvalores de la matriz A
eigen1, eigen2 = la.eig(A)
print("eigenvectores:")
print(eigen2)
print("Eigenvalores", eigen1)

# La traza:
p = eigen1[0] + eigen1[1]
q = eigen1[0]*eigen1[1]

print("Traza(A) =", p)

# Discriminante:
print("Discriminante(A) =", p**2 - 4*q)
```

```python
#PC CENTRO

# Definimos la función solución:
def S(c1, c2, t):
  x1 = c1*np.cos(t) + c2*np.sin(t)
  x2 = c2*np.cos(t) - c1*np.sin(t)
  x = [x1, x2]
  return np.array(x)

# Se definen las constantes

t = np.linspace(-20, 20, 6000)
c1 = np.linspace(-25, 25, 10)
c2 = np.linspace(20, 100, 10)

# Soluciones

plt.figure(figsize=(8, 8))
for i, j in zip(c1, c2):
  sol = np.array(S(i,j,t))
  plt.plot(sol[0,:],sol[1,:])
plt.title("Soluciones particulares del sistema de ecuaciones")
plt.xlabel("x(t)")
plt.ylabel("y(t)")
```

# **Ejercicio 3.** 

\begin{eqnarray*}
\frac{d^2 x}{d t^2} + \omega_0^2 x & = & 0, \ \omega_0 > 0  \\
\end{eqnarray*}

```python
# En este caso dv/dt = -omega_0^2*x, por tanto la matriz quedará:
# Se elige, antes, un valor para omega_0
w = 5 #omega_0
# parametros de la matriz y posteriormente matriz A
a = 0
b = 1
c = -w**2
d = 0
A = np.array([[a, b],[c, d]])

#Eigen todo
eigen1, eigen2 = la.eig(A)
print("eigenvectores:")
print(eigen2)
print("Eigenvalores", eigen1)

# La traza:
p = eigen1[0] + eigen1[1]
q = eigen1[0]*eigen1[1]

print("Traza(A) =", p)

# Discriminante:
print("Discriminante(A) =", p**2 - 4*q)
```

```python
#CENTRO TAMBOR
# Soluciones del tipo:
def S(c1, c2, t):
  x1 = c1*np.cos(w*t) + c2*np.sin(w*t)
  x2 = c2*w*np.cos(w*t) - c1*w*np.sin(w*t)
  x = [x1, x2]
  return np.array(x)

# Constantes varias

t = np.linspace(-25, 25, 4500)
c1 = np.linspace(-25, 25, 10)
c2 = np.linspace(20, 100, 10)

# Grafica de las sol

plt.figure(figsize=(10,10))
for i, j in zip(c1, c2):
  sol = np.array(S(i,j,t))
  plt.plot(sol[0,:],sol[1,:])
plt.title("Soluciones particulares del sistema de ecuaciones")
plt.xlim(-200, 200)
plt.ylim(-200, 200)
plt.xlabel("x(t)")
plt.ylabel("y(t)")
```

# **Ejercicio 4.** 

\begin{eqnarray*}
\frac{d x}{d t} & = & -2 x  \\
\frac{d y}{dt } & = &  2 z  \\
\frac{d z}{dt } & = &  - 2 y  \\
\end{eqnarray*}

```python
# Dado el sistema, ahora la matriz será 3x3;
A = np.array([[-2,0,0],[0,0,2],[0,-2,0]])

#Eigen todo
eigen1, eigen2 = la.eig(A)
print("eigenvectores:")
print(eigen2)
print("Eigenvalores", eigen1)

# La traza:
p = eigen1[0] + eigen1[1]
q = eigen1[0]*eigen1[1]

print("Traza(A) =", p)

# Discriminante:
print("Discriminante(A) =", p**2 - 4*q)
```

```python
# Nuevamente, punto critco centro
# Las soluciones serán del tipo:
def S(c1, c2, c3, t):
  x1 = c1*np.exp(-2*t)
  x2 = c2*np.cos(2*t) + c3*np.sin(2*t)
  x3 = -c2*np.sin(2*t) + c3*np.cos(2*t)
  x = [x1, x2, x3]
  return np.array(x)

# Ctes

t = np.linspace(-50, 50, 5000)
c1 = np.linspace(-25, 25, 10)
c2 = np.linspace(-25, 25, 10)
c3 = np.linspace(-25, 25, 10)

# Grafica de soluciones

plt.figure()
ax = plt.axes(projection='3d')
for i, j, k in zip(c1, c2, c3):
  sol = np.array(S(i,j,k,t))
  ax.plot3D(sol[0,:], sol[1,:], sol[2,:])
plt.title("Soluciones particulares del sistema de ecuaciones")
ax.set_xlabel("x(t)")
ax.set_ylabel("y(t)")
ax.set_zlabel("z(t)")
```

# **Ejercicio 5.** 

\begin{eqnarray*}
\frac{d x}{d t} & = & - x + z  \\
\frac{d y}{dt } & = &  3 y  \\
\frac{d z}{dt } & = &  - x - z \\
\end{eqnarray*}

```python
# Dado el sistema, se obtiene que la matriz A será:
A = np.array([[-1,0,1],[0,3,0],[-1,0,-1]])

#Eigen todo
eigen1, eigen2 = la.eig(A)
print("eigenvectores:")
print(eigen2)
print("Eigenvalores", eigen1)

# La traza:
p = eigen1[0] + eigen1[1]
q = eigen1[0]*eigen1[1]

print("Traza(A) =", p)

# Discriminante:
print("Discriminante(A) =", p**2 - 4*q)
```

```python
# Solcuiones del tipo
def S(c1, c2, c3, t):
  x1 = np.exp(-t)*(c1*np.cos(t) + c3*np.sin(t))
  x2 = c2*np.exp(3*t)
  x3 = np.exp(-t)*(c3*np.cos(t) - c1*np.sin(t))
  x = [x1, x2, x3]
  return np.array(x)

# Ctes
t = np.linspace(-3, 3, 5000)
c1 = np.linspace(-25, 25, 10)
c2 = np.linspace(-25, 25, 10)
c3 = np.linspace(-10, 10, 10)

# Grafica de las soluciones
plt.figure()
ax = plt.axes(projection='3d')
for i, j, k in zip(c1, c2, c3):
  sol = np.array(S(i,j,k,t))
  ax.plot3D(sol[0,:], sol[1,:], sol[2,:])
plt.title("Soluciones particulares del sistema de ecuaciones")
ax.set_xlabel("x(t)")
ax.set_ylabel("y(t)")
ax.set_zlabel("z(t)")
```

```python
# Para el sistema, la matriz A:
A = np.array([[-1, 0],[1, 2]])
#Eigen todo
eigen1, eigen2 = la.eig(A)
print("eigenvectores:")
print(eigen2)
print("Eigenvalores", eigen1)

# La traza:
p = eigen1[0] + eigen1[1]
q = eigen1[0]*eigen1[1]

print("Traza(A) =", p)

# Discriminante:
print("Discriminante(A) =", p**2 - 4*q)
```

# **Ejercicio 10:** 

Se tiene el siguiente sistema de resortes acoplados con dos masas.  

Screen Shot 2021-03-18 at 00.35.32.png


El sistema esta sujeto del primer resorte de la izquierda. Cuando el sistema se deja en reposo, las longitudes de los resortes son $L_1$ y $L_2$.

Las ecuaciones de movimiento estan dadas por las siguientes ecuaciones diferenciales acopladas:

\begin{eqnarray*}
m_1 x_1^{''} + b_1 x_1^{'}+k_1 (x_1-L_1)-k_2(x_2-x_1-L_2) & = & 0 \\
m_2 x_2^{''} + b_2 x_2^{'}+k_2(x_2-x_1-L_2) & = & 0 \\
\end{eqnarray*}

Encuentre las soluciones como funciones de $t$ y grafíquelas, así como las trayectorias en el espacio fase.

Suponga que $L_1=L_2=1$ y que las masas son iguales $m_1=m_2=1$.  

**10.1)** Encuentre los eigenvalores del sistema y diga como son las soluciones.  

**10.2)** Caso sin fricción. $b_1=b_2=0$, $k_1=6$, $k_2=4$, condiciones iniciales $(x_1(0), x_1^{'}(0), x_2(0), x_2^{'}(0))=(1,0,4,0)$ 

**10.3)** Igual que el caso anterior pero, con fricción: $b_1=0.1, b_2=0.2$

```python
# Definiendo la funcion para el oscilador

def sol(s, t, k1, k2, b1, b2):
        x1, v1, x2, v2 = s
        dsdt = [v1, -x1*(k1+k2) -b1*v1+k2*x2+k1-k2, v2, k2*x1-k2*x2-b2*v2+k2] 
        return dsdt

# Definimos las condiciones iniciales:
s0 = [1.2, 0, 3.0, 0]
```

```python
# Podemos decir que la parte real de los eigen valores tiene valor cero.
# Se trata de un error de redondeo
# Por ende, no podemos afirmar nada sobre la categoria del punto critico
# tiempo de integración
t = np.linspace(0, 30, 7000)

# Integrando con scipy
s = odeint(sol, s0, t, args=(k1, k2, b1, b2))


plt.figure(figsize=(8,10))
plt.plot(t, s[:, 0], "b", label="x1(t)")
plt.plot(t, s[:, 2], "g", label="x2(t)")
plt.legend(loc="best")
plt.title("Soluciones de x1(t) y x2(t) con respecto al tiempo")
plt.xlabel("t")
plt.ylabel("")
plt.grid()
plt.show()

plt.figure(figsize=(8,10))
plt.plot(s[:, 0], s[:, 2], "b")
plt.legend(loc="best")
plt.title("Soluciones de x1(t) y x2(t) en espacio fase")
plt.xlabel("x1(t)")
plt.ylabel("x2(t)")
plt.grid()
plt.show()
```
